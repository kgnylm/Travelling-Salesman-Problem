```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gurobipy as gp
from gurobipy import GRB
from gurobipy import quicksum
from gurobipy import Model
from VRPlib import *
```

```python
df_dist = pd.read_excel('real_distances_30_customers.xlsx',sheet_name = "distance", header=None)
df_req=pd.read_excel('real_distances_30_customers.xlsx',sheet_name="demand")
df_cap=pd.read_excel('real_distances_30_customers.xlsx',sheet_name="Vehicle_Capacity",header=None)
customer_requirements=df_req.values.tolist()
vehicle_capacity=df_cap.values.tolist()
distance_matrix = df_dist.values.tolist()
nodes = list(range(df_dist.shape[0]))
```

```python
routes_saving=savingsAlgorithm(distance_matrix,120,nodes[1:],nodes)
for i in routes_saving:
    i.insert(0,0)
    i.append(0)
distance_Saving=[]
distance_2opt=[]
distance_2ex=[]
routes_2ex =[]
routes_2opt=[]
for route in routes_saving:
    distance_Saving.append(routeDistance(distance_matrix,route))
```

```python
routes_2ex=twoExchange(distance_matrix,routes_saving,0,1,customer_requirements)
routes_2ex.append(routes_saving[-1])

for route in routes_2ex:
    route_dist_ex=routeDistance(distance_matrix,route)
    distance_2ex.append(route_dist_ex)
    new_route,new_length=two_opt(route,route_dist_ex,distance_matrix)
    routes_2opt.append(new_route)
    distance_2opt.append(new_length)
```

```python
print("Distance from saving algorithm : {} ".format(distance_Saving))
print("Distance from 2-ex algorithm : {}".format(distance_2ex))
print("Distance from 2-opt algorithm : {}".format(distance_2opt))

print("Routes from saving algorithm : {} ".format(routes_saving))
print("Routes from 2-ex algorithm : {}".format(routes_2ex))
print("Routes from 2-opt algorithm : {}".format(routes_2opt))
```

```python
vrp = Model("vrp_model")
vrp.setParam("TimeLimit", 30)
#2445
#m.setParam('OutputFlag', False)

#Decision variables
x = vrp.addVars(nodes,nodes,lb=0, vtype=GRB.BINARY, name="x")
u = vrp.addVars(nodes,lb=0, vtype=GRB.CONTINUOUS, name="u")

#Objective function
vrp.setObjective(quicksum(distance_matrix[i][j]*x[i,j] for i in range(len(distance_matrix)) for j in range(len(distance_matrix))), gp.GRB.MINIMIZE)
vehicle_capacity=120
#demand=[]
"""
for i in range(30):
    temp=[]
    for j in range(30):
        temp.append(customer_requirements[i][1])
    demand.append(temp)
"""
#Add constraint
vrp.addConstrs(((quicksum(x[i,j] for j in nodes if j != i) == 1) for i in nodes if i != 0), name = '1');
vrp.addConstrs(((quicksum(x[i,j] for i in nodes if i != j) == 1) for j in nodes if j != 0), name = '2');
vrp.addConstr(((quicksum(x[0,j] for j in nodes) == 3)), name = '3');
vrp.addConstr(((quicksum(x[i,0] for i in nodes) == 3)), name = '4');
vrp.addConstrs(((u[i] - u[j] + vehicle_capacity*x[i,j] <= vehicle_capacity - customer_requirements[j][1]) 
for i in nodes for j in nodes if i!=j and i != 0 and j != 0), name = '5');
vrp.addConstrs(((u[i] <= vehicle_capacity) for i in nodes), name = '6');
vrp.addConstrs(((u[i] >= customer_requirements[i][1]) for i in nodes), name = '7');

#Solve
vrp.optimize()

#Print solution
print("Optimal solution:")
for v in vrp.getVars():
    if v.x > 0:
        print(v.varName, v.x)

print("Optimal objective value:", vrp.objVal)

```
